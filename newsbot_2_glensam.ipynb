# 🧠 NewsBot Intelligence System 2.0
A complete NLP pipeline with classification, sentiment, summarization, translation, and chatbot interface.

```python
# Install required packages (run once)
!pip install transformers sentence-transformers scikit-learn nltk deep-translator

import nltk
nltk.download('punkt')

from transformers import pipeline
from sentence_transformers import SentenceTransformer, util
from deep_translator import GoogleTranslator
```

```python
# Sample news dataset
news_data = [
    {"text": "AI is transforming the tech industry with new innovations.", "category": "technology"},
    {"text": "The stock market saw a significant drop today due to inflation concerns.", "category": "business"},
    {"text": "The local team won their championship game in a stunning victory.", "category": "sports"}
]

texts = [item["text"] for item in news_data]
texts
```

```python
# Sentiment Analysis
sentiment_model = pipeline("sentiment-analysis")

for text in texts:
    print(text)
    print(sentiment_model(text))
    print("-"*50)
```

```python
# Summarization
summarizer = pipeline("text-generation", model="t5-small")

for text in texts:
    summary = summarizer(text, max_length=30, min_length=10, do_sample=False)
    print("Original:", text)
    print("Summary:", summary[0]['generated_text'])
    print("-"*50)
```

```python
# Semantic Search
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(texts, convert_to_tensor=True)

query = "technology advancements"
query_embedding = model.encode(query, convert_to_tensor=True)

scores = util.cos_sim(query_embedding, embeddings)

for i, score in enumerate(scores[0]):
    print(f"Score: {score:.4f} | Text: {texts[i]}")
```

```python
# Translation
for text in texts:
    translated = GoogleTranslator(source='auto', target='es').translate(text)
    print("Original:", text)
    print("Spanish:", translated)
    print("-"*50)
```

```python
# Simple Chatbot Interface

def chatbot(query):
    if "summary" in query.lower():
        return summarizer(texts[0])[0]['generated_text']
    elif "sentiment" in query.lower():
        return sentiment_model(texts[0])
    elif "translate" in query.lower():
        return GoogleTranslator(source='auto', target='fr').translate(texts[0])
    else:
        return "I can summarize, analyze sentiment, or translate news."

# Test chatbot
print(chatbot("give me summary"))
print(chatbot("sentiment"))
print(chatbot("translate"))
```
